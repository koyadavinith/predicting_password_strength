```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('password_strength.csv',error_bad_lines = False)
data.head()
```

```python
data.strength.unique()
```

```python
data.isna().sum()
```

```python
data[data['password'].isna()]
```

```python
data.dropna(inplace = True)
```

```python
sns.countplot(data['strength'])
```

**To Provide robustness to our model we have to shuffle the data**

```python
password_tuple = np.array(data)
```

```python
import random
random.shuffle(password_tuple)
```

```python
x = [label[0] for label in password_tuple]
y = [label[1] for label in password_tuple]
```

```python
x
```

```python
def word_to_char(inputs):
    return list(inputs)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer= word_to_char)
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector = X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df = pd.DataFrame(first_document_vector.T.todense(),index = vectorizer.get_feature_names(),columns = ['TF-IDF'])
df
```

```python
df.sort_values(by = 'TF-IDF',ascending = False)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(multi_class= 'multinomial')
lr.fit(x_train,y_train)
```

**Example**

```python
temp = np.array(['Jagityal&123'])
pred = vectorizer.transform(temp)
lr.predict(pred)
```

```python
preds = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, auc, confusion_matrix,f1_score
print(f"Accuracy score is {accuracy_score(y_test, preds)}")
print(f"Confusion matrix is \n{confusion_matrix(y_test, preds)}")
print(f"f1 score  is \n{f1_score(y_test, preds,average = 'weighted')}")
```
